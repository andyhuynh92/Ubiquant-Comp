Benchmarking between polars and pandas

```python
import pandas as pd
import polars as pl
import numpy as np
import gc
import time
```

```python
%%time
df = pl.read_parquet('parquet/train_low_mem.parquet')
df2 = pl.read_parquet('stockclose.parquet')
```

```python
df2 = pl.DataFrame({['time_id'].dtype}
```

```python
%%time
temp = []
t = time.localtime()
current_time = time.strftime("%H:%M:%S", t)
print(current_time)
for j in range(0,300):
    if j % 50 == 0:
        print('Starting correlation with feature',j)
    temp = []
    string = 'f_'+str(j)
    tempdf = (df[['time_id', 'target', string]]
            .groupby(['time_id'])
            .agg([pl.pearson_corr(string, 'target')])
            .sort(by='time_id')
           )
    #tempdf = pl.DataFrame({'investment_id':range(0,3774), string:temp})
    if j ==0:
        corrdf = tempdf
    else:
        corrdf = corrdf.join(tempdf, on='time_id')

t = time.localtime()
current_time = time.strftime("%H:%M:%S", t)
print(current_time)
```
